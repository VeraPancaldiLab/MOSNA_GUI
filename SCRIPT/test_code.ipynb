```python
import pandas as pd 
import glob
from pathlib import Path
```

```python
object_data = glob.glob("../pandas_data/*.csv")
files = dict()

for file in object_data:
    obj = pd.read_parquet(Path(file).with_suffix('.parquet'))
    files.setdefault(Path(file).name, obj)
```
